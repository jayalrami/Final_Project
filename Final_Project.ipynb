```python

import pandas as pd

class Stock_Market:
    def __init__(self):
        self.df = None

    # ================= DATA LOADING =================
    def load_dataset(self):
        try:
            self.df = pd.read_csv("StockMarket.csv")
            print("Dataset loaded successfully!")
        except Exception as e:
            print("Error loading dataset:", e)
    
    # ================= EXPLORE DATA =================
    def explore_data(self):
        if self.df is None:
            print("Please load the dataset first!")
            return
        
        while True:
            print("\n1. First 5 rows\n2. Last 5 rows\n3. Column names\n4. index\n5. Data types\n6. Basic info\n7. Exit")
            ch = int(input("Enter your choice: "))
            
            if ch == 1:
                print(self.df.head())
            elif ch == 2:
                print(self.df.tail())
            elif ch == 3:
                print(self.df.columns)
            elif ch == 4:
                print(self.df.index)    
            elif ch == 5:
                print(self.df.dtypes)
            elif ch == 6:
                self.df.info()  
            elif ch == 7:
                print("Exit...")
                break
            else:
                print("Invalid choice!")

analyzer = Stock_Market()
analyzer.load_dataset()
analyzer.explore_data()
```

# MISSING DATA

```python
import pandas as pd

class Stock_Market:
    def __init__(self):
        self.df = None

    # ================= DATA LOADING =================
    def load_dataset(self):
        try:
            self.df = pd.read_csv("StockMarket.csv")
            print("Dataset loaded successfully!")
        except Exception as e:
            print("Error loading dataset:", e)

    # ================= MISSING DATA =================        
    def handle_missing_data(self):
        if self.df is None:
            print(" Please load the dataset first!")
            return
        
        while True:
            print("\n1. Display missing rows\n2. Fill with mean\n3. Drop missing rows\n4. Exit")
            ch = int(input("Enter your choice: "))

            if ch == 1:
                print("\nMissing Rows:")
                print(self.df[self.df.isnull().any(axis=1)])

            elif ch == 2:
                self.df.fillna(self.df.mean(numeric_only=True), inplace=True)
                print(" Missing values filled with mean!")

            elif ch == 3:
                self.df.dropna(inplace=True)
                print(" Missing rows dropped!")

            elif ch == 4:
                print("Exit missing data menu...")
                break

            else:
                print(" Invalid choice!")
analyzer = Stock_Market()
analyzer.load_dataset()
analyzer.handle_missing_data()
    
```

# VISUALIZATION

```python
#### import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
      
class Stock_Market:
    def __init__(self):
        self.df = None

    # ================= DATA LOADING =================
    def load_dataset(self):
        try:
            self.df = pd.read_csv("StockMarket.csv")
            print("Dataset loaded successfully!")
        except Exception as e:
            print("Error:", e)
    
    # ================= VISUALIZATION =================
    def visualize(self):
        if self.df is None:
            print("Please load the dataset first!")
            return
        
        # Convert Date to datetime and extract Year
        if 'Date' in self.df.columns:
            self.df['Date'] = pd.to_datetime(self.df['Date'])
            self.df['Year'] = self.df['Date'].dt.year        
        
        while True:   
            print("""
1. Bar Plot (Average High Price by Year)
2. Line Plot (Open Price Trend by Year)
3. Scatter Plot (Open vs High)
4. Pie Chart (Total Price by Year)
5. Histogram (Price Distribution)
6. Stack Plot (Open vs High vs Low by Year)
7. Exit
            """)
            
            ch = int(input("Enter your choice: "))

            if ch == 1:
                data = self.df.groupby('Year')['High'].mean()
                data.plot(kind='bar')
                plt.title("Average High Price by Year")
                plt.ylabel("High Price")
                plt.show()

            elif ch == 2:
                data = self.df.groupby('Year')['Open'].mean()
                data.plot(kind='line')
                plt.title("Open Price Trend by Year")
                plt.ylabel("Open Price")
                plt.show()

            elif ch == 3:
                plt.scatter(self.df['Open'], self.df['High'])
                plt.xlabel("Open Price")
                plt.ylabel("High Price")
                plt.title("Open vs High")
                plt.show()

            elif ch == 4:
                data = self.df.groupby('Year')['Price'].sum()
                data.plot(kind='pie', autopct='%1.1f%%')
                plt.title("Total Price by Year")
                plt.ylabel("")
                plt.show()

            elif ch == 5:
                plt.hist(self.df['Price'], bins=10)
                plt.title("Price Distribution")
                plt.xlabel("Price")
                plt.ylabel("Frequency")
                plt.show()

            elif ch == 6:
                yearly_data = self.df.groupby('Year')[['Open', 'High', 'Low']].mean()
                plt.stackplot(yearly_data.index,
                              yearly_data['Open'],
                              yearly_data['High'],
                              yearly_data['Low'],
                              labels=['Open', 'High', 'Low'])
                plt.title("Stack Plot (Open vs High vs Low by Year)")
                plt.legend()
                plt.show()

            elif ch == 7:
                print("Exit.....")
                break  

            else:
                print(" Invalid choice!")

analyzer = Stock_Market()
analyzer.load_dataset()
analyzer.visualize()
```

# SEABORN

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

class Stock_Market:
    def __init__(self):
        self.df = None

    # ================= DATA LOADING =================
    def load_dataset(self):
        try:
            self.df = pd.read_csv("StockMarket.csv")
            print("Dataset loaded successfully!")
        except Exception as e:
            print("Error:", e)
    
    # ================= SEABORN VISUALIZATION =================
    def seaborn_visuals(self):
        if self.df is None:
            print(" Please load the dataset first!")
            return
        
        # Convert Date to datetime and extract Year
        if 'Date' in self.df.columns:
            self.df['Date'] = pd.to_datetime(self.df['Date'])
            self.df['Year'] = self.df['Date'].dt.year
        
        # Boxplot
        sns.boxplot(x='Year', y='Price', data=self.df)
        plt.title("Price by Year")
        plt.show()

        # Heatmap (Correlation)
        corr = self.df.corr(numeric_only=True)
        sns.heatmap(corr, annot=True, cmap='coolwarm')
        plt.title("Correlation Heatmap")
        plt.show()

analyzer = Stock_Market()
analyzer.load_dataset()
analyzer.seaborn_visuals()
```
